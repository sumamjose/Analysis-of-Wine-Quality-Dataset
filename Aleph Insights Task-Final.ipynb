## Measures of Center and Spread

Measures of Center includes Mean, Median and Mode & Measures of Spread includes Range, IQR (InterQuartile Range), Variance and Standard Deviation.

```python
# Calculate the median, range, quantiles, IQR and variance for the variables in the dataset rounded to 4 decimal places.
table1 = []
for i in range(0, len(dataset.columns)):
    median = np.round(dataset.iloc[:,i].median(), decimals=4)
    variable_range = np.round(max(dataset.iloc[:,i]) - min(dataset.iloc[:,i]), decimals=4)
    quantiles = np.round([dataset.iloc[:,i].quantile(0),
                 dataset.iloc[:,i].quantile(0.25),
                 dataset.iloc[:,i].quantile(0.50),
                 dataset.iloc[:,i].quantile(0.75),
                 dataset.iloc[:,i].quantile(1)], decimals=4)
    IQR = np.round(dataset.iloc[:,i].quantile(0.75) - dataset.iloc[:,i].quantile(0.25), decimals=4)
    variance = np.round(dataset.iloc[:,i].var(), decimals=4)
    table1.append([dataset.columns[i], median, variable_range, quantiles, IQR, variance])
print(tabulate(table1, headers=['Variable', 'Median', 'Range', 'Quantiles', 'IQR', 'Variance']))
```

# TASK 1.2 - Clustering

Clustering is the process of segregating similar data points into clusters based on their similarity measure.

## K-Means Clustering

### Clustering without outlier treatment

```python
# Import the required libraries.
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
```

```python
# Remove the categoric target variable, 'quality' from the dataset before performing clustering.
dataset_new = dataset[dataset.columns[0:11]]
dataset_new.head()
```

```python
# Check for missing values if any.
print ("The total number of missing values in the dataset is", dataset.isna().sum().sum())
```

There are no missing values in the dataset.

```python
# Normalize the variables in the dataset to convert them to the same scale.
dataset_scaled = RobustScaler().fit_transform(dataset_new)
dataset_scaled = pd.DataFrame(dataset_scaled, columns=dataset_new.columns)
dataset_scaled
```

```python
# Function to obtain principal components and to plot them graphically using a scatter plot.
def pca(dataset_PCA):
    PCA_features = PCA(n_components=2).fit_transform(dataset_PCA)
    plt.figure(figsize=(8, 6))
    plt.scatter(PCA_features[:,0],PCA_features[:,1])
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.title('Principal components of data points')
    plt.show()
    return PCA_features
# Create a copy of the scaled dataset.
dataset_scaled_1 = dataset_scaled.copy()
# Obtain two principal components from the 11 variables using dimensionality reduction and obtain a scatter plot.
PCA_features_obtained_1 = pca(dataset_scaled_1)
```

```python
# Function for Elbow method using WCSS that employs K-Means clustering.
def elbow_method_kmeans(PCA_features):
    WCSS = []
    # Experiment KMeans clustering with different number of clusters ranging from 1 to 10.
    for i in range(1, 11):
        kmeans_clustering = KMeans(n_clusters = i)
        kmeans_clustering.fit(PCA_features)
        # Calculate the WCSS value
        WCSS.append(kmeans_clustering.inertia_)
    # Plot the WCSS values obtained for different number of clusters (1 to 10) to decide the ideal number of clusters.
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, 11), WCSS)
    plt.title('WCSS obtained for different number of clusters')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS obtained')
    plt.show()
# Decide the number of clusters for K-Means clustering using the Elbow method that employs WCSS.
elbow_method_kmeans(PCA_features_obtained_1)
```

Elbow point is obtained at number of clusters equal to 3. Hence use 3 clusters in the clustering model.

```python
# Function for clustering model.
def clustering_model(PCA_features, cluster_model):
    # Fit the model to the data and perform prediction of clusters.
    clusters = cluster_model.fit_predict(PCA_features)
    return clusters
# Set up a K-Means clustering model with 3 clusters.
cluster_model_1 = KMeans(n_clusters=3, init='k-means++', n_init=100, max_iter=1000)
# Call the function to perform fitting and prediction using K-Means clustering.
clusters_obtained_1 = clustering_model(PCA_features_obtained_1, cluster_model_1)
```

```python
# Function to plot the clusters obtained.
def cluster_plot(clusters, PCA_features, color_dict):
    colors = [color_dict[x] for x in clusters]
    plt.figure(figsize=(8, 6))
    for sample in range(len(clusters)):
        plt.scatter(PCA_features[sample][0], PCA_features[sample][1], color = colors[sample])
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.title('Clusters obtained')
    plt.show()
# Define the colour dictionary.
color_dict_1 = {0:'blue', 1:'green', 2:'red'}
# Call the function to plot the clusters obtained using K-Means clustering.
cluster_plot(clusters_obtained_1, PCA_features_obtained_1, color_dict_1)
```

The clusters are indicated in different colours above.

```python
# Calculate the Silhouette Coefficinet.
print ('Silhouette score obtained for K-Means clustering is: ', silhouette_score(PCA_features_obtained_1, clusters_obtained_1, metric='euclidean'))
```

### Clustering with outlier treatment

#### Capping and Flooring method to treat outliers:

```python
# Function for treating outliers.
def outlier_removal(dataset_outliers):
    # Deal with the outliers identified graphically in the variables of the dataset using the Capping and Flooring method.
    for i in range(0, len(dataset_outliers.columns)):
        Percentiles = dataset_outliers.iloc[:,i].quantile([0.01, 0.99]).values
        dataset_outliers.iloc[:,i] = np.clip(dataset_outliers.iloc[:,i], Percentiles[0], Percentiles[1])
    return dataset_outliers
# Select only the variables with extreme outliers and pass it to the function to remove the outliers in them.
data_outlier = dataset_new[['residual sugar', 'chlorides', 'sulphates']].copy()
dataset_outliers = outlier_removal(data_outlier)
dataset_outliers.head()
```

```python
# Replace the outlier values with the outlier treated values in the original dataset.
dataset_new_copy = dataset_new.copy()
dataset_new_copy[['residual sugar', 'chlorides', 'sulphates']] = dataset_outliers[['residual sugar', 'chlorides', 'sulphates']].values
dataset_new_copy
```

Outliers are replaced with their 99 percentile and 1 percentile values by Capping and Flooring method.

```python
# Call the function to re-plot boxplots for the variables in the dataset after dealing with the outliers.
box_plot(dataset_outliers) 
```

Extreme outliers were removed from the variables of the dataset.

```python
# Normalize the variables in the dataset to convert them to the same scale.
dataset_scaled_copy = RobustScaler().fit_transform(dataset_new_copy)
dataset_scaled_copy = pd.DataFrame(dataset_scaled_copy, columns=dataset_new_copy.columns)
dataset_scaled_copy
```

```python
# Call the function to obtain principal components and to plot them graphically.
PCA_features_obtained_2 = pca(dataset_scaled_copy)
```

```python
# Call the function to find the number of clusters using Elbow method (WCSS)
elbow_method_kmeans(PCA_features_obtained_2)
```

Elbow point is obtained at number of clusters equal to 3, same as with outliers. Hence use 3 clusters in the clustering model.

```python
# Call the function to perform clustering to obtain 3 clusters.
clusters_obtained_2 = clustering_model(PCA_features_obtained_2, cluster_model_1)
```

```python
# Call the function to plot the clusters obtained.
cluster_plot(clusters_obtained_2, PCA_features_obtained_2, color_dict_1)
```

```python
# Calculate the Silhouette Coefficinet.
print ('Silhouette score obtained for K-Means clustering with outlier treatment is: ', silhouette_score(PCA_features_obtained_2, clusters_obtained_2, metric='euclidean'))
```

From what is observed, outlier removal does not cause any improvement in the clustering process. The silhouette score slightly decreased by around 0.035 with outlier removal.

# TASK 1.3 - Multi-Class Classification

Multi-Class classification is the process of classifying the data points to different target classes based on the feature values. The target variable in the wine quality dataset is 'quality', which is a discrete variable with 6 classes ranging from 3 to 8.

```python
# Import the required libraries.
from sklearn.compose import ColumnTransformer
from imblearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from imblearn.over_sampling import SMOTE
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB 
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import GridSearchCV
```

```python
# Divide the dataset into features, X and target, y.
y = dataset['quality']
X = dataset[dataset.columns[0:11]]

# Perform oversampling to resolve the target class imbalance.
oversample = SMOTE(k_neighbors=4)
X_oversampled, y_oversampled = oversample.fit_resample(X,y)

# Plot the target variable distribution after oversampling.
dataset_oversampled = pd.concat([X_oversampled, y_oversampled], axis=1)
target_distribution(dataset_oversampled)

# Select the numeric features in the dataset.
numeric_features = X.select_dtypes([np.number]).columns
numeric_features
```

The target variable, 'quality' is now balanced after SMOTE oversampling.

This indicates that all the features in X are numeric. Hence feature scaling could be performed on all the features in the dataset. Since there are no categoric variables in the dataset, encoding is not required.

## Train-Test Split

```python
# Split the data into training set and test set in the ratio of 4:1.
X_train, X_test, y_train, y_test = train_test_split(X_oversampled, y_oversampled, test_size=0.20, random_state=0, stratify=y_oversampled)
# To print the size of Train and Test sets
print ("The size of X and y of train set is:\n", X_train.shape, y_train.shape)  
print ("The size of X and y of test set is:\n", X_test.shape, y_test.shape)     
```

## Model Comparison using Cross Validation Score

```python
# Define the dictionary of models to be compared.
models_used = []
models_used.append(('Decision Tree', DecisionTreeClassifier()))
models_used.append(('Random Forest', RandomForestClassifier()))
models_used.append(('Gradient Boosting', GradientBoostingClassifier()))
models_used.append(('K-Nearest Neighbor', KNeighborsClassifier()))
models_used.append(('Gaussian Naive Bayes', GaussianNB()))
models_used.append(('Logistic Regression', LogisticRegression(multi_class='ovr')))
models_used.append(('Support Vector Machine', SVC(decision_function_shape='ovo')))

# Take copy of X and y of train set.
Xtrain_copy = X_train.copy()
ytrain_copy = y_train.copy()
for model_name, model in models_used:
    # Create the pipeline for model training that includes feature scaling and feature selection.
    pipeline = Pipeline(steps=[('scaler1', RobustScaler(with_centering=False, with_scaling=True)), 
                               ('feature_selection', SelectKBest(score_func=chi2, k=10)),
                               ('model', model)])    
    # Determine the cross validation scores for the model.
    score = cross_val_score(pipeline, Xtrain_copy, ytrain_copy, scoring='accuracy', cv=10, n_jobs=-1)
    print ('{}: '.format(model_name))
    print ('Accuracy score is {}'.format(score.mean()))
```

Random Forest classifier had the highest cross validation accuracy score of around 86 %. Hence the model to be used here is Random Forest.

## Training and Prediction using Random Forest Classifier

```python
# Take copy of X and y of train set and test set.
X_train1 = X_train.copy()
y_train1 = y_train.copy()
X_test1 = X_test.copy()
y_test1 = y_test.copy()

# Define the pipeline that employs Random Forest model to be used for training, that includes feature scaling and feature selection.
pipeline = Pipeline(steps=[('scaler', RobustScaler(with_centering=False, with_scaling=True)), 
                           ('feature_selection', SelectKBest(score_func=chi2, k=10)),
                           ('model', RandomForestClassifier(class_weight='balanced_subsample'))])
# Train the model with the training data.
random_forest = pipeline.fit(X_train1, y_train1)
# Make predictions on the test data using the trained model.
y_pred1 = random_forest.predict(X_test1)
```

## Evaluate Model Performance

```python
# Assess the model performance.
def model_performance(y_test, y_pred, prob):
    print('Accuracy:', accuracy_score(y_test, y_pred))
    print('Precision:', precision_score(y_test, y_pred, average='macro'))
    print('Recall:', recall_score(y_test, y_pred, average='macro'))
    print('F1-score:', f1_score(y_test, y_pred, average='macro'))
    # Calculate ROC_AUC score.
    print('ROC_AUC score:', roc_auc_score(y_test, prob, multi_class='ovr'))
    print('Classification Report:\n', classification_report(y_test,y_pred))
    print('Confusion Matrix:\n', confusion_matrix(y_test,y_pred))  
# Prediction of probability.
prob1 = random_forest.predict_proba(X_test1)
model_performance(y_test1, y_pred1, prob1)
```

The different performance scores obtained using the Random Forest model are indicated above. From the scores obtained, the model performs reasonably well. The accuracy obtained is around 83.79 %.

## Hyperparameter Tuning of Random Forest Classifier using Grid Search

A grid of different hyperparameters are provided, from which the best set of parameters is estimated for the model used. These are then used in the prediction of results.

```python
# Perform hyperparameter tuning of the Random Forest classifier to improve the performance.
# Take copy of the train and test sets.
X_train2 = X_train.copy()
y_train2 = y_train.copy()
X_test2 = X_test.copy()
y_test2 = y_test.copy()
# Define the grid of parameters of the Random Forest model to evaluate.
grid_values = {'model__n_estimators': [100, 200, 300, 400, 500, 1000], 
               'model__criterion': ['gini','entropy'],
               'model__min_samples_split': [2,5,10], 
               'model__min_samples_leaf': [1,5,10]}
# Initialize Grid Search CV.
grid_search = GridSearchCV(pipeline, grid_values, cv = 10, scoring='accuracy') 
# Training using Grid Search applied to the Random forest classifier.
grid_search.fit(X_train2, y_train2)
# Prediction using the best estimates obtained using Grid Search.
y_pred2 = grid_search.best_estimator_.predict(X_test2)
```

## Evaluate Model Performance with Hyperparameter Tuning

```python
# Assess the model performance with hyperparameter tuning.
# Prediction of probability.
prob2 = grid_search.best_estimator_.predict_proba(X_test2)
model_performance(y_test2, y_pred2, prob2)
# Plot the Confusion Matrix.
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test2, y_pred2), display_labels=grid_search.classes_).plot()
plt.title('Confusion matrix of Random Forest classifier with Grid search')
plt.show()
```

The Random Forest classifier with hyperparameter tuning obtained an accuracy of around 84.82 % and has helped to improve the correct predictions in the Confusion Matrix slightly. However, the model was less efficient in predicting the majority wine quality classes of '5' and '6', which has to be investigated further.

Seven machine learning classifier models were compared, which includes Decision Tree (DT), Random Forest (RF), Gradient Boosting (GB), K-Nearest Neighbor (KNN), Logistic Regression (LR), Gaussian Naive Bayes (GNB) and Support Vector Machines (SVM). Random Forest showed the best performance at an accuracy score of around 86 %, followed by GB. This was followed by DT and KNN, which showed similar results. This was followed by LR and SVM. However, GNB showed a poor performance at slightly over 50 %. The model which was finally used in modelling and prediction was Random Forest classifier. Hyperparameter Tuning using the Grid Search algorithm further improved the model's performance.

### Advantages and Disadvantages of the Random Forest Classifier

Advantages:
- A Random Forest classifier can deal with both classification and regression analysis.
- It could deal with different types of features, both continuous and categoric.
- It is an ensemble based model and hence combines the results from different decision trees. Hence helps to reduce overfitting to a greater extent. Thus the accuracy gets improved.
- Treating of missing values and scaling of features is not a necessary step in Random Forest.
- It has a range of hyperparameters that could be tuned to further improve the accuracy of the model.

Disadvantages:
- It takes more time to run as it contains multiple decision trees. Computational power and training time required is high.
- It is less interpretable as it contains several trees.
